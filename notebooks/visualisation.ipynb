```python
from huggingface_hub import login
import os
from datasets import load_dataset

hf_token = os.environ['HUGGINGFACE_TOKEN']
login(token=hf_token)
```

```python
ds = load_dataset("imageomics/sentinel-beetles", split="train", streaming=True)
inp = []
for i, example in enumerate(ds):
    if i > 20: break
    new = {}
    new['relative_img']=example['file_path']
    new['colorpicker_img']=example['colorpicker_full_path']
    new['scalebar_img']=example['scalebar_full_path']
    new['scientificName']=example['scientificName']
    new['domainID']=example['domainID']
    inp.append(new)
```
